```python
#1
#finding the mean and standard deviation for the real attributes
mean = df.mean()
std = df.std()

#z-normalization is (xi-mean)/standard deviation
normalized_df = (df - mean) / std
normalized_df
```

```python
import numpy as np
import pandas as pd

#using np.array to make the normalized_data to an array format
data = np.array(normalized_df)

#calculating the mean using np.mean
mean_vector = np.mean(data, axis=0)

# centereing the data is nothing but xi-mean
centered_data = data - mean_vector

#the number of data points, n
n = centered_data.shape[0]

#to store the covariance matrix, initialized an empty cov_matrix
cov_matrix = np.zeros((centered_data.shape[1], centered_data.shape[1]))

#covariance matrix without using np.outer
# normalizing by n but not N-1 as mentioned in the code
for i in range(centered_data.shape[1]):
    for j in range(centered_data.shape[1]):
        cov_matrix[i, j] = np.sum(centered_data[:, i] * centered_data[:, j]) / n  

print("Sample Covariance Matrix (Σ):")
print(cov_matrix)
```

```python
#calculating the covariance matrix with the help of np.cov
cov_matrix_numpy = np.cov(centered_data, rowvar=False, bias=True)
cov_matrix_numpy
```

```python
#conditionizing to check both the values are not correct or not
are_equal = np.allclose(cov_matrix, cov_matrix_numpy, rtol=1e-6)

if are_equal:
    print("The manually calculated covariance matrix matches the numpy.cov result.")
else:
    print("The matrices do not match.")
```

```python
#question 3
#considered a threshold, the difference between xi and xi+1 to be minimum so that we know the dominant vector
threshold = 0.000001
```

```python
# we will always choose the x0 vector randmoly
d = cov_matrix.shape[0]
x0 = np.random.rand(d)
x0
```

```python
#xi and xi-1
#it becomed the previous one and the one we intend to calculate as xi
x_prev = np.zeros(d)
x_prev
```

```python
#euclidean norm 
def euclidean_norm(vector):
    return np.sqrt(np.sum(vector**2))
```

```python
#to achieve convergence, we are iterating the function
while euclidean_norm(x0 - x_prev) > threshold:
    x_prev = x0
    x0 = np.dot(cov_matrix, x0)
    
    # the maximum absolute value
    max_abs_value = np.max(np.abs(x0))
    max_index = np.argmax(np.abs(x0))
    
    # scaling the vector to have the largest value as 1
    x0 /= max_abs_value

#the eigenvector to have unit length, I normalized it
x0 /= np.linalg.norm(x0)

#the ratio of consecutive elements, yields to the largest eigen value
largest_eigenvalue = x0[max_index] / x_prev[max_index]

print("Dominant Eigenvalue:", largest_eigenvalue)
print("Dominant Eigenvector:", x0)
```

```python
#question h
#using the df components

def pca(df, n_components=None):
    
    # Mean-center the data
    X_mean = np.mean(df, axis=0)
    X_centered = df - X_mean

    #covariance matrix
    covariance_matrix = np.cov(X_centered, rowvar=False)

    #eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eigh(covariance_matrix)

    #ordering the eigenvalues and eigenvectors in descending order
    sorted_indices = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[sorted_indices]
    eigenvectors = eigenvectors[:, sorted_indices]

    #if n_components is specified, selecting the top components
    if n_components is not None:
        eigenvectors = eigenvectors[:, :n_components]

    # Project the data onto the new coordinate system
    transformed_data = np.dot(X_centered, eigenvectors)

    return transformed_data, eigenvalues, eigenvectors
```

```python
# Perform PCA to retain 95% of variance
transformed_data, eigenvalues, eigenvectors = pca(df)

# number of components needed to preserve 95% of variance
total_variance = np.sum(eigenvalues)
cumulative_variance = np.cumsum(eigenvalues) / total_variance
n_components_to_preserve_95_variance = np.argmax(cumulative_variance >= 0.95) + 1

print(f"Number of components to preserve 95% variance: {n_components_to_preserve_95_variance}")

#using the first n_components_to_preserve_95_variance eigenvectors as the new basis vectors
new_basis_vectors = eigenvectors[:, :n_components_to_preserve_95_variance]

# the coordinates of the first 10 data points using the new basis vectors
data_points = df[:10]
coordinates = np.dot(data_points - np.mean(df, axis=0), new_basis_vectors)
print("Coordinates of the first 10 data points in the new basis:")
print(coordinates)

```
